# c

```python
def binomial_pmf(N, p):
    height = []

    for i in range(N+1):
        height.append(comb(N, i) * (p**i) * ((1-p)**(N-i)))
    
    return height
```

```python
from math import comb

def plot_bionomial(ax, N, p):
    height = binomial_pmf(N, p)

    ax.bar(range(N+1), height)
    ax.title.set_text(f"PMF N={N} p = {p}")    
    plt.setp(ax, xlabel="Valeus")
    plt.setp(ax, ylabel="Probability")  
```

```python
fig, ax = plt.subplots(1, 3)
plot_bionomial(ax[0], 5, 0.2)
plot_bionomial(ax[1], 10, 0.3)
plot_bionomial(ax[2], 20, 0.1)

fig.set_size_inches(17,5)
plt.show()
```

```python
p = 0.2
likelihoods = []
for N in range(1, 21):
    pmf = binomial_pmf(N,p)
    
    likelihood = 1
    for sample in samples:
        if sample >= len(pmf):
            likelihood = 0
        else:
            likelihood *= pmf[sample]
    likelihoods.append(likelihood)

plt.bar(range(1, 21), likelihoods)
plt.title("Likelihood")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("N", fontsize=18)
plt.ylabel('Probability', fontsize=16)
plt.show()
```

```python
p = 0.2
likelihoods = []
for N in range(1, 21):
    pmf = binomial_pmf(N,p)
    
    likelihood = 0
    flag = False
    for sample in samples:
        if sample >= len(pmf):
            flag = True
        else:
            likelihood += math.log(pmf[sample])
    
    if flag:
        likelihoods.append(0)    
    else:
        likelihoods.append(likelihood)

plt.bar(range(1, 21), likelihoods)
plt.title("Likelihood")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("N", fontsize=18)
plt.ylabel('Probability', fontsize=16)
plt.show()
```

# j

```python
def log_likelihood(samples, a, b):
    n = len(samples)
    return n * math.log(a) + n * math.log(b) + (a-1)*(sum([math.log(x) for x in samples])) + (b-1)*sum([math.log(1-x**a) for x in samples]) 
```

```python
def Kumaraswamy(a, b, x):
    return a * b * (x ** (a-1)) * ((1 - x**a) ** (b-1)) 
```

# j

```python
maxim_p = 0
best_param = None
for a in range(1, 13):
    for b in range(1, 13):
        p = log_likelihood(samples, a, b)
        if p > maxim_p:
            maxim_p = p
            best_param = (a, b)

print(best_param)
```

```python
def evolution_log_likelihood(param):
    n = len(samples)
    a = param[0]
    b = param[1]
    
    try:
        output = n * math.log(a) + n * math.log(b) + (a-1)*(sum([math.log(x) for x in samples])) + (b-1)*sum([math.log(1-x**a) for x in samples])
    except ValueError:
        output = float("-inf")
    
    return -1*output
```

```python
from scipy.optimize import differential_evolution

bounds = [(0, 12), (0, 12)]
result = differential_evolution(evolution_log_likelihood, bounds)
print(result.x)
```

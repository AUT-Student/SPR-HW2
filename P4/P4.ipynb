# Libraries

```python
# Import Allowed Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
samples = [0, 2, 0, 3, 1, 2, 2, 2, 0, 2, 3, 1, 0, 0, 2, 3, 3, 2, 3, 0]
```

# a

```python
height = np.zeros(4)
for sample in samples:
    height[sample] += 1

plt.bar(range(4), height)
plt.title("Number occurance of value of samples")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("Value", fontsize=18)
plt.ylabel('Occurrences', fontsize=16)
plt.show()
```

# b

```python
height = np.zeros(4)
for sample in samples:
    height[sample] += 1

for i in range(4):
    height[i] /= len(samples)
    
plt.bar(range(4), height)
plt.title("Probability Mass Function of Samples")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("Value", fontsize=18)
plt.ylabel('Probability', fontsize=16)
plt.show()
```

# c

```python
def binomial_pmf(N, p):
    height = []

    for i in range(N+1):
        height.append(comb(N, i) * (p**i) * ((1-p)**(N-i)))
    
    return height
```

```python
from math import comb

def plot_bionomial(ax, N, p):
    height = binomial_pmf(N, p)

    ax.bar(range(N+1), height)
    ax.title.set_text(f"PMF N={N} p = {p}")    
    plt.setp(ax, xlabel="Valeus")
    plt.setp(ax, ylabel="Probability")  
```

```python
fig, ax = plt.subplots(1, 3)
plot_bionomial(ax[0], 5, 0.2)
plot_bionomial(ax[1], 10, 0.3)
plot_bionomial(ax[2], 20, 0.1)

fig.set_size_inches(17,5)
plt.show()
```

# d

```python
p = 0.2
likelihoods = []
for N in range(1, 21):
    pmf = binomial_pmf(N,p)
    
    likelihood = 1
    for sample in samples:
        if sample >= len(pmf):
            likelihood = 0
        else:
            likelihood *= pmf[sample]
    likelihoods.append(likelihood)

plt.bar(range(1, 21), likelihoods)
plt.title("Likelihood")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("N", fontsize=18)
plt.ylabel('Probability', fontsize=16)
plt.show()
```

# e

```python
p = 0.2
likelihoods = []
for N in range(1, 21):
    pmf = binomial_pmf(N,p)
    
    likelihood = 0
    flag = False
    for sample in samples:
        if sample >= len(pmf):
            flag = True
        else:
            likelihood += math.log(pmf[sample])
    
    if flag:
        likelihoods.append(0)    
    else:
        likelihoods.append(likelihood)

plt.bar(range(1, 21), likelihoods)
plt.title("Likelihood")

loc, labels = plt.xticks()
plt.xticks(np.arange(0, max(loc), step=1))

plt.xlabel("N", fontsize=18)
plt.ylabel('Probability', fontsize=16)
plt.show()
```

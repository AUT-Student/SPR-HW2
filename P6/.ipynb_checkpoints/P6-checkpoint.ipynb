```python
# Import Allowed Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
import glob
from PIL import Image
from sklearn.metrics import accuracy_score 
```

# a

```python
def all_path(part, setnumber, raw):
    if part=="train":
        if raw:
            return glob.glob(f"./pratheepan/train/set{setnumber}/images/*")
        else:
            return glob.glob(f"./pratheepan/train/set{setnumber}/groundtruth/*.png")
    elif part=="test":
        if raw:
            return glob.glob(f"./pratheepan/test/images/*.jpg")
        else:
            return glob.glob(f"./pratheepan/test/groundtruth/*.png")
```

```python
def load_img_data(path):
    """ Load an image and convert it to its data"""
    im = Image.open(path)
    data = np.array([[im.getpixel((i, j)) for i in range(im.size[0])] for j in range(im.size[1])])
    return data
```

```python
priors = [0, 0]

for path in all_path("train", 1, False):
    img_data = load_img_data(path)
    
    for pixel_array in img_data:
        for pixel in pixel_array:
            if pixel[0] == 0:
                priors[0] += 1
            else:
                priors[1] += 1

all_pixels = priors[0] + priors[1]
priors = [priors[0]/all_pixels, priors[1]/all_pixels]
```

# b

```python
class_data = [[], []]
raw_data_path = all_path("train", 1, True)
label_data_path = all_path("train", 1, False)

for path_index in range(len(raw_data_path)):
    raw_img_data = load_img_data(raw_data_path[path_index])
    label_img_data = load_img_data(label_data_path[path_index])
    
    for i, pixel_array in enumerate(raw_img_data):
        for j, pixel in enumerate(pixel_array):
            if label_img_data[i][j][0] == 0:
                class_data[0].append(pixel)
            else:
                class_data[1].append(pixel)
```

```python
class_mean = []
class_cov = []

for class_ in [0, 1]:
    class_data_df = []
    for row in class_data[class_]:
        class_data_df.append({'R': row[0], 'G': row[1], 'B': row[2]})
   
    class_data_df = pd.DataFrame(class_data_df)
    class_mean.append(class_data_df.mean())
    class_cov.append(class_data_df.cov())
```

# c

```python
def p_likelihood(class_, x):
    mean = class_mean[class_].values
    sigma = class_cov[class_].values
    sigma_det = np.linalg.det(sigma)
    sigma_inv = np.linalg.inv(sigma)

    n = 3
    part2 = np.exp(-0.5*np.sum((x-mean.T).dot(sigma_inv)*(x-mean.T),axis=1))
    
    p = 1/((2*np.pi*sigma_det)**(n/2)) * np.exp(-0.5*np.sum((x-mean.T).dot(sigma_inv)*(x-mean.T),axis=1))
    
    return p
```

```python
def predict(x_array):
    likelihood_0 = p_likelihood(0, x_array)
    likelihood_1 = p_likelihood(1, x_array)
    
    output = []
    for i in range(len(x_array)):
        output.append(likelihood_0[i] * priors[0] < likelihood_1[i] * priors[1])
    return  output
```

```python
def display_classified_image(image_path):
    output = []
    img_data = load_img_data(image_path)
    for pixel_array in img_data:
        output_array = []

        for pr in predict(pixel_array):
            if pr:
                output_array.append(255)
            else:
                output_array.append(0)

        output.append(output_array)

    img = Image.fromarray(np.uint8(output) , 'L')
    display(img)        
```

# d

```python
def mdc_predict(x):
    m0 = class_mean[0].values
    m1 = class_mean[1].values
    
    return np.linalg.norm(m0 - np.array(x)) > np.linalg.norm(m1 - np.array(x))
```

```python
def display_image_mdc(image_path):
    output = []
    img_data = load_img_data(image_path)
    for pixel_array in img_data:
        output_array = []

        for pixel in pixel_array:
            if mdc_predict(pixel):
                output_array.append(255)
            else:
                output_array.append(0)

        output.append(output_array)

    img = Image.fromarray(np.uint8(output) , 'L')
    display(img) 
```

# e

```python
raw_data_path = all_path("test", 1, True)
label_data_path = all_path("test", 1, False)

true_list = []
predict_list = []

for path_index in range(len(raw_data_path)):
    raw_img_data = load_img_data(raw_data_path[path_index])
    label_img_data = load_img_data(label_data_path[path_index])
    
    output = []
    for pixel_array in raw_img_data:
        output_array = []

        for pr in predict(pixel_array):
            if pr:
                output_array.append(255)
            else:
                output_array.append(0)

        output.append(output_array)

    for i in range(len(output)):
        for j in range(len(output[0])):
            true_list.append(label_img_data[i][j][0])
            predict_list.append(output[i][j])
            

print(accuracy_score(true_list, predict_list))
```
